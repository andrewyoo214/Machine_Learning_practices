## 1. Read & Check the given biofeedback data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO
bio1 = pd.read_csv('F:/files/RESEARCH/Datasets/DT Platform/biofeedback_1/bio_practice.csv')
```

#### 1) EMG(근전도) - time domain

```python
plt.figure(figsize=(20,10))
plt.plot(bio1.Time, bio1.EMG) #EMG = 근전도검사
plt.xlabel('duration_time')
plt.ylabel('EMG')
plt.title('ELECTROMYOGRAPHY', fontsize=20)
timecoords = [600.0625, 1500.0625, 900.0625, 1200.0625] #10분, 25분, 15분, 20분 간격표시
colors=['r','r','m','m']
for xc,c in zip(timecoords, colors):
    plt.axvline(x=xc, c=c, linestyle='--', linewidth=3)
plt.show()
```

#### 2) Skin Conductance (피부전도도) - time domain

```python
plt.figure(figsize=(20,10))
plt.plot(bio1.Time, bio1.Skin_Cond) #Skin Conductance = 피부전도도
plt.xlabel('duration_time')
plt.ylabel('skin conductance')
plt.title('SKIN CONDUCTANCE', fontsize=20)
timecoords = [600.0625, 1500.0625, 900.0625, 1200.0625] #10분, 25분, 15분, 20분 간격표시
colors=['r','r','m','m']
for xc,c in zip(timecoords, colors):
    plt.axvline(x=xc, c=c, linestyle='--', linewidth=3)
plt.show()
```

#### 3) Temperature(체온) - time domain

```python
plt.figure(figsize=(20,10))
plt.plot(bio1.Time, bio1.Temp) #Temp = 체온
plt.xlabel('duration_time')
plt.ylabel('temperature')
plt.title('TEMPERATURE', fontsize=20)
timecoords = [600.0625, 1500.0625, 900.0625, 1200.0625] #10분, 25분, 15분, 20분 간격표시
colors=['r','r','m','m']
for xc,c in zip(timecoords, colors):
    plt.axvline(x=xc, c=c, linestyle='--', linewidth=3)
plt.show()
```

#### 4) EEG- Theta, Alpha Amples(뇌파) - time domain

```python
plt.figure(figsize=(20,10))
plt.plot(bio1.Time, bio1.Theta_Ampl, label='theta') #Theta Ample = 뇌파 - 세타파
plt.plot(bio1.Time, bio1.Alpha_Ampl,'green', label='alpha') #Alpha Ample = 뇌파 - 알파파
plt.xlabel('duration_time')
plt.ylabel('EEG')
plt.title('EEG- THETA and ALPHA', fontsize=20)
timecoords = [600.0625, 1500.0625, 900.0625, 1200.0625] #10분, 25분, 15분, 20분 간격표시
colors=['r','r','m','m']
for xc,c in zip(timecoords, colors):
    plt.axvline(x=xc, c=c, linestyle='--', linewidth=3)
plt.show()
```
